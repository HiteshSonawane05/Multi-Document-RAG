```python
from PyPDF2 import PdfReader
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
os.listdir()
```

```python
text = ""
```

```python
for file in os.listdir():
    if file.endswith(".pdf"):

        file_obj = PdfReader(file)

        for page in file_obj.pages:
            text += page.extract_text() + "\n"
```

```python
#print(text)
```

```python
len(text)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap =200)
```

```python
chunks = text_splitter.split_text(text)
```

```python
len(chunks)
```

```python
print(chunks[-1])
```

```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.get_or_create_collection(name='multi_pdf_vdb')
collection.add(
    ids = ['id_'+str(i) for i in range(len(chunks))],
    documents=chunks)
```

```python
search_results = collection.query(query_texts=['What is a Epochs?'], n_results=3)
```

```python
search_results['documents'][0][2]
```

```python
llm = ChatGroq(model = 'llama-3.3-70b-versatile')
```

```python
question = input('Enter Question:')
search_results = collection.query(query_texts=[question], n_results=1)
prompt = f'''You are an assistant for question-answering tasks. Use the following pieces of retrieved context to answer the question. If you don't know the answer, just say that you don't know. Use three sentences maximum and keep the answer concise.
Question: {question} 
Context: {search_results['documents'][0]} '''
```

```python
# print(prompt)
```

```python
response = llm.invoke(prompt)
```
